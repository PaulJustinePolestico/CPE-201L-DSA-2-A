```python
from collections import deque
import time

class Graph:
    def __init__(self):
        self.adj_list = {}

    def add_vertex(self, vertex):
        if vertex not in self.adj_list:

            self.adj_list[vertex] = []

    def add_edge(self, vertex1, vertex2, directed=False):
        self.add_vertex(vertex1)
        self.add_vertex(vertex2)

        # Add the edge
        self.adj_list[vertex1].append(vertex2)


        self.adj_list[vertex1].sort()

        if not directed:
            self.adj_list[vertex2].append(vertex1)
            self.adj_list[vertex2].sort()

    def display(self):
        print("Adjacency List:")
        for vertex, neighbors in self.adj_list.items():
            print(f"{vertex}: {neighbors}")


def dfs_recursive(graph, start, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    if start not in visited:
        visited.add(start)
        path.append(start)


        for neighbor in graph.adj_list.get(start, []):
            if neighbor not in visited:
                dfs_recursive(graph, neighbor, visited, path)

    return path


def bfs(graph, start, visited):
    queue = deque([start])
    path = []

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            visited.add(vertex)
            path.append(vertex)


            for neighbor in graph.adj_list.get(vertex, []):
                if neighbor not in visited:
                    queue.append(neighbor)

    return path



G26 = Graph()


for v in ['q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']:
    G26.add_vertex(v)


G26.add_edge('q', 's', directed=True)
G26.add_edge('q', 't', directed=True)
G26.add_edge('q', 'y', directed=True)
G26.add_edge('s', 'v', directed=True)
G26.add_edge('s', 'w', directed=True)
G26.add_edge('t', 'x', directed=True)
G26.add_edge('t', 'y', directed=True)
G26.add_edge('v', 'w', directed=True)
G26.add_edge('x', 'x', directed=True)
G26.add_edge('x', 'z', directed=True)
G26.add_edge('z', 'x', directed=True)


G26.add_edge('r', 'u', directed=True)
G26.add_edge('u', 'y', directed=True)


G26.display()
print("-" * 25)




print("--- BFS Traversal ---")
all_visited_bfs = set()
bfs_path_q = bfs(G26, 'q', all_visited_bfs)
bfs_path_r = bfs(G26, 'r', all_visited_bfs)
print(f"BFS Full Traversal Order: {bfs_path_q + bfs_path_r}")


print("\n--- DFS Traversal ---")
all_visited_dfs = set()
dfs_path_q = dfs_recursive(G26, 'q', all_visited_dfs, [])
dfs_path_r = dfs_recursive(G26, 'r', all_visited_dfs, [])
print(f"DFS Full Traversal Order: {dfs_path_q + dfs_path_r}")
```
