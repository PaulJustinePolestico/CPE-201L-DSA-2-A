```python
#Array of even integers
from array import array

def traverse_array(arr):
  print("Traversing the array:")
  for i,element in enumerate(arr):
    print(f"Arr[{i}]: {element}")


def reverse_array(arr):
  arr.reverse()
  print("Array reversed:", arr)

def reverse_array_elements(arr):
  arr.reverse()
  print("Array elemenyts in reverse order:")
  for i,element in enumerate(arr):
    print(f"Arr[{i}]: {element}")



def maximum_element(arr):
  max_element = max(arr)
  print(f"The maximum element in the array is: {max_element}")


def main():
  arr = array('i', [i for i in range(20,50,2)])


  while True:
    print("\nMenu:")
    print("1. Display the Original Array")
    print("2. Traverse the array")
    print("3. Reverse the array elements")
    print("4. Reverse the Array")
    print("5. Maximum element in the array")
    print("6. Exit")
    choice = input("Enter which operation you want to perform: ")

    if choice == '1':
      print("Array of Even Integers:", arr)
    elif choice == '2':
      traverse_array(arr)
    elif choice == '3':
      reverse_array_elements(arr)
    elif choice == '4':
      reverse_array(arr)
    elif choice == '5':
      maximum_element(arr)
    elif choice == '6':
      print("Exiting the program.")
      break


if __name__ == "__main__":
  main()
```
