```python
def traverse_array(arr):
  """Function to traverse and print all elements in the array"""
  print("Traversing the array:")
  for i,element in enumerate(arr):
    print(f"Index{i}: {element}")

#Reverse Function
def reverse(arr):
  """Function to reverse the array"""
  arr.reverse()
  print("Array reversed:", arr)

#Sort Function
def sort_array(arr):
  """Function to sort the array"""
  arr.sort()
  print("Array sorted:", arr)

#Insert Function
def insert_into_array(arr):
  """Function to insert an element into the array"""
  try:
    index = int(input("Enter the index to insert at: "))
    value = int(input("Enter the value to insert: "))
    if 0 <= index <= len(arr):
      arr.insert(index, value)
      print("Array after insertion:", arr)
    else:
      print("Invalid index.")
  except ValueError:
    print("Invalid input. Please enter integers for index and value.")


#Delete Function
def delete_from_array(arr):
  """Function to delete an element from the array"""
  try:
    index = int(input("Enter the index of the element to delete: "))
    if 0 <= index < len(arr):
      del arr[index]
      print("Array after deletion:", arr)
    else:
      print("Invalid index.")
  except ValueError:
    print("Invalid input. Please enter an integer for the index.")


#Modify Function
def modify_array(arr):
  """Function to modify an element in the array"""
  try:
    index = int(input("Enter the index of the element to modify: "))
    value = int(input("Enter the new value: "))
    if 0 <= index < len(arr):
      arr[index] = value
      print("Array modified:", arr)
    else:
      print("Invalid index.")
  except ValueError:
    print("Invalid input. Please enter integers for index and value.")

#Append Function
def append_to_array(arr):
  """Function to append an element to the array"""
  try:
    value = int(input("Enter the value to append: "))
    arr.append(value)
    print("Array after appending: ", arr)
  except ValueError:
    print("Invalid input. Please enter an integer for the value.")

#Main Function
def main():
  #Initialize an array
  arr = [10,20,30,40,50]
  print("Original array:", arr)

  while True:
    print("\nMenu:")
    print("1. Traverse the array")
    print("2. Reverse the array")
    print("3. Sort the Array")
    print("4. Modify the Array")
    print("5. Insert into the Array")
    print("6. Delete from the Array")
    print("7. Append the Array")
    print("8. Exit")
    choice = input("Enter your choice: ")

    if choice == '1':
      traverse_array(arr)
    elif choice == '2':
      reverse(arr)
    elif choice == '3':
     sort_array(arr)
    elif choice == '4':
      modify_array(arr)
    elif choice == '5':
      insert_into_array(arr)
    elif choice == '6':
      delete_from_array(arr)
    elif choice == '7':
      append_to_array(arr)
    elif choice == '8':
      print("Exiting the program.")
      break

if __name__ == "__main__":
  main()
```
